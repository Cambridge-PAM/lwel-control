```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import rc_context
from matplotlib import rcParams
import matplotlib.cm as cm

file_storage=r'C:\Users\tl457\OneDrive - University Of Cambridge 1\3_Code\lwel-control\test220531.h5'
f = h5py.File(file_storage, 'r')

# specify matplotlib configuration file name
rc_fname = r'C:\Users\tl457\Documents\GitHub\spectroscopy\plotting_params.txt'
```

```python
list(f.keys())
```

```python
f.attrs['creation_time']
```

```python
grp = f['dataset_1']
print(list(grp.keys()))
dset1=grp['image_1']
dset2=grp['spec_1']
print(grp.attrs['timestamp'])
print(dset1)
print(dset2)
```

```python
pixel_size=3.45 # um
x=np.linspace(0,np.shape(dset1)[1]*pixel_size,np.shape(dset1)[1])
y=np.linspace(0,np.shape(dset1)[0]*pixel_size,np.shape(dset1)[0])
z=dset1
with rc_context(fname=rc_fname):
    plt.pcolormesh(x,y,z,shading='gouraud')
    plt.xlabel('X / $\mu$m')
    plt.ylabel('Y / $\mu$m')
```

```python
with rc_context(fname=rc_fname):
    plt.plot(dset2[0],dset2[1])
    plt.xlabel('Wavelength / nm')
    plt.ylabel('Counts / a.u.')
    plt.xlim(350,850)
    plt.ylim(0)
```
